take input from the user

```python
From = int(input("From: "))  
To = int(input("To: "))
```

find the sum of the cube of each digit

```python
print("Armstrong numbers within {0} and {1} are :".format(From,To))
for num in range(From,To + 1):
   sum = 0  
   temp = num  
   while temp > 0:  
       digit = temp % 10  
       sum += digit ** 3  
       temp //= 10  
       if num == sum:  
            print(num)
```
